# Data Exploration with Matplotlib

```python
# import the library
import matplotlib.pyplot as plt
```

### 1. การสร้างแผนภูมิเส้น (Line charts)

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]    # X-AXIS
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.5, 10289.7, 14958.3] # Y-AXIS

# พล็อตแผนภูมิเส้น
plt.plot(
    years,        # แกน x 
    gdp,          # แกน y
    color='green', # กำหนดสีเส้น
    marker='o',    # กำหนดรูปแบบมาร์เกอร์ของจุดข้อมูล
    linestyle='solid' # กำหนดรูปแบบเส้น
)

# เพิ่มชื่อแผนภูมิ
plt.title("Nominal GDP")

# เพิ่มป้ายกำกับของแกน x
plt.xlabel("Years")

# เพิ่มป้ายกำกับของแกน y
plt.ylabel("Billions of $")

# แสดงกราฟ
plt.show()
```

### 2. การสร้างแผนภูมิแท่ง (Bar charts)

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

xs = [i + 0.1 for i, _ in enumerate(movies)]

# พล็อตแผนภูมิแท่ง
plt.bar(
    xs,        # แกน x
    num_oscars # แกน y
)

# เพิ่มป้ายกำกับแกน y
plt.ylabel("# of Academy Awards")

# เพ่มชื่อแผนภูมิ
plt.title("My Favorite Movies")

# เพิ่มชื่อแท่งข้อมูล
plt.xticks(xs, movies)

# แสดงกราฟ
plt.show()
```

### 3. ฮิสโตแกรม (Histograms)

```python
from collections import Counter

grades = [83, 95, 91, 87, 70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar(
    [x - 4 for x in histogram.keys()],  # แกน x
    histogram.values(),                 # แกน y
    color='green',                      # สีแท่งข้อมูล
    width=8,                            # ความกว้างของแท่งข้อมูล
)

# กำหนดช่วงค่าของแกน x และแกน y
plt.axis([-10, 105, 0, 5])    


# แท่งข้อมูลแต่ละแท่งแทนคะแนน 0, 10, 20, ...
plt.xticks([10 * i for i in range(11)])


# ป้ายกำกับแกน x
plt.xlabel("Decile")

# ป้ายกำกับแกน y
plt.ylabel("# of Students")

# เพิ่มชื่อแผนภาพ
plt.title("Distribution of Exam 1 Grades")

# แสดงแผนภาพ
plt.show()
```

### 4. แผนภูมิแบบกระจาย (Scatterplots)

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(
        label,
        xy=(friend_count, minute_count),  # วางป้ายกำกับข้อมูลไว้
        xytext=(5, -5),                   # เยื้องลงมาทางขวาล่างของจุดข้อมูล
        textcoords='offset points'        # ระยะป้ายกำกับแบบ offset
    )

# เพิ่มชื่อแผนภูมิ
plt.title("Daily Minutes vs. Number of Friends")

# เพิ่มป้ายกำกับแกน x
plt.xlabel("# of Friends")

# เพิ่มป้ายกำกับแกน y
plt.ylabel("Daily minutes spent on the site")

# แสดงแผนภาพ
plt.show()
```
