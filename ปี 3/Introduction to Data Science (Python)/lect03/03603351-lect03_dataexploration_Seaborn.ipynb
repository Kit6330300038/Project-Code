# Data Exploration with Seaborn

## ไลบารี่ Seaborn

Seaborn (https://seaborn.pydata.org/) เป็นไลบารี่ภาษาไพธอนที่ต่อยอดมาจาก Matplotlib เราสามารถใช้ Seaborn สร้างกราฟฟิกทางสถิติที่มีทั้งคุณภาพและความสวยงาม ได้อย่างง่ายดาย

```python
# import the library
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(
    context='notebook',          # context: paper, notebook, talk, poster
    style='white',               # style: white, dark, whitegrid, darkgrid, ticks
)                             
```

### 1. การสร้างแผนภูมิเส้น (Line charts)

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.5, 10289.7, 14958.3]

# สร้าง dataframe
df = pd.DataFrame({'Years': years, 'Billions of $': gdp})

# พล็อตแผนภูมิเส้น
sns.lineplot(
    x='Years',   # แกน x 
    y='Billions of $',     # แกน y
    data=df,
    marker='o',
    color='g'
)

plt.title('Nominal GDP')
```

### 2. การสร้างแผนภูมิแท่ง (Bar charts)

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

df = pd.DataFrame({'Movies': movies, '# of Academy Awards': num_oscars})

sns.barplot(
    x='Movies',
    y='# of Academy Awards',
    data=df
)

plt.title('My Favorite Movies')
```

### 3. ฮิสโตแกรม (Histograms)

```python
grades = [83, 95, 91, 87, 70,0,85,82,100,67,73,77,0]

sns.histplot(grades, bins=10, kde=False)

# ป้ายกำกับแกน x
plt.xlabel("Decile")

# ป้ายกำกับแกน y
plt.ylabel("# of Students")

# เพิ่มชื่อแผนภาพ
plt.title("Distribution of Exam 1 Grades")
```

### 4. แผนภูมิแบบกระจาย (Scatterplots)

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

df = pd.DataFrame({'# of Friends': friends, 
                   'Daily minutes spent on the site': minutes})

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(
        label,
        xy=(friend_count, minute_count),  # วางป้ายกำกับข้อมูลไว้
        xytext=(5, -5),                   # เยื้องลงมาทางขวาล่างของจุดข้อมูล
        textcoords='offset points'        # ระยะป้ายกำกับแบบ offset
    )
    
sns.scatterplot(
    x='# of Friends',
    y='Daily minutes spent on the site',
    data=df,
)
```
